# **Caso práctico bancario end-to-end**

## 1. Objetivo del caso

Queremos cerrar el workshop con una visión integradora.

La idea es mostrar cómo diferentes técnicas pueden convivir dentro de un flujo bancario real.

## 2. Escenario planteado

Supongamos una entidad financiera que recibe tres tipos de información:

1. **Datos tabulares del cliente**  
   - ingresos
   - saldo
   - productos
   - retrasos
   - endeudamiento

2. **Texto libre**  
   - email del cliente
   - observaciones del gestor
   - descripción de incidencia

3. **Imagen o documento visual**  
   - justificante
   - documento de identidad
   - comprobante
   - formulario escaneado

## 3. Objetivo del sistema

Construir un flujo que ayude a:
- priorizar riesgo,
- detectar fraude documental,
- clasificar incidencias,
- y apoyar a negocio sin sustituir la revisión humana.

## 4. Arquitectura conceptual end-to-end

### Bloque tabular
Modelo clásico:
- logística
- random forest
- boosting

### Bloque de texto
Modelo NLP:
- fine-tuning con Hugging Face

### Bloque visual
Modelo de imagen:
- CNN
- o transfer learning en visión

### Capa de decisión final
- reglas de negocio
- score combinado
- revisión manual de casos críticos

## 5. Flujo operativo

1. entra una solicitud
2. se calcula un score tabular
3. se analiza el texto asociado
4. se valida el documento o imagen
5. se combinan señales
6. se genera una recomendación:
   - aprobar
   - revisar
   - rechazar

```python
import pandas as pd
import numpy as np

# Ejemplo conceptual de scores parciales
df_scores = pd.DataFrame({
    "cliente_id": [1, 2, 3, 4],
    "score_tabular": [0.82, 0.35, 0.61, 0.12],
    "score_texto_fraude": [0.15, 0.80, 0.30, 0.10],
    "score_documento_sospechoso": [0.05, 0.70, 0.20, 0.15]
})

display(df_scores)
```

## 6. Regla de agregación simple

En la práctica real esto puede hacerse de muchas formas:
- modelo meta
- reglas expertas
- scorecard combinada
- sistema híbrido

Aquí mostramos una regla sencilla y muy explicable.

```python
df_scores["score_total_riesgo"] = (
    0.5 * df_scores["score_tabular"] +
    0.3 * df_scores["score_texto_fraude"] +
    0.2 * df_scores["score_documento_sospechoso"]
)

def decision(score):
    if score >= 0.70:
        return "rechazar / revisar urgente"
    elif score >= 0.40:
        return "revisión manual"
    else:
        return "aprobación automática"

df_scores["decision"] = df_scores["score_total_riesgo"].apply(decision)
display(df_scores)
```

## 7. Qué gana el banco con este enfoque

### Ventajas
- más información por caso
- mejor priorización
- más robustez frente a fraude complejo
- integración de señales heterogéneas

### Riesgos
- más complejidad operacional
- más exigencia de trazabilidad
- necesidad de gobernanza fuerte
- posible conflicto entre rendimiento e interpretabilidad

## 8. Qué debería exigir auditoría / validación

Un sistema así no puede aprobarse solo porque “funciona bien”.

Habría que revisar:
- calidad de datos,
- leakage,
- drift,
- sesgo,
- umbrales,
- documentación,
- estabilidad,
- explicabilidad,
- y protocolo de fallback humano.

## 9. Qué se puede enseñar aquí como cierre del workshop

no existe un único modelo mágico.

En la práctica moderna, muchas veces el valor está en:
- combinar modelos,

- y construir sistemas trazables.
